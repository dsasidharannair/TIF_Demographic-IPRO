```python
import json
import pandas as pd
import plotly.express as px
import geopandas as gpd
import plotly.graph_objects as go
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
tif_transfers_file = pd.read_csv('/content/drive/480-497Demographic TIF Team/Transfers/DataFolder/Programmed_Transfers/TIF_Transfers_merged.csv')
tif_bounds = gpd.read_file('Boundaries - Tax Increment Financing Districts.geojson')
```

# All Transfers 2017-2026

```python
tif_transfers_dict = {}
for index, row in tif_transfers_file.iterrows():
    if row['TIF'] in tif_transfers_dict:
        tif_transfers_dict[row['TIF']] += int(row['Amount'])
    else:
        tif_transfers_dict[row['TIF']] = int(row['Amount'])

tif_transfers = pd.DataFrame([(k,v) for k,v in tif_transfers_dict.items()], columns=['TIF', 'Amount'])
tif_bounds.rename(columns={"name":"TIF"}, inplace=True)
merged = tif_bounds.merge(tif_transfers, on='TIF', how='left')
merged
```

```python
# Create a base layer for the interactive heatmap for census tracts
fig = px.choropleth_mapbox(
    merged,
    geojson=merged.geometry,
    locations=merged.index,
    color='Amount',  # Color based on population
    color_continuous_scale=["blue", "red"],  # Choose a color scale
    # range_color=[min(tif_transfers['Amount']), max(tif_transfers['Amount'])],  # Set color range explicitly
    range_color=[-7000000, 14000000],  # Set color range explicitly
    mapbox_style="carto-positron",
    center={"lat": merged.centroid.y.mean(), "lon": merged.centroid.x.mean()},
    zoom=10,
    opacity=0.5,
    labels={'TIF'},
    hover_name='TIF',  # Display community area name on hover
)
fig.show()
```

# Transfers By Year (20127 - 2026)

```python
for year in range(2017, 2027):
    tif_transfers_dict = {}
    for index, row in tif_transfers_file[tif_transfers_file['Year'] == year].iterrows():
        if row['TIF'] in tif_transfers_dict:
            tif_transfers_dict[row['TIF']] += int(row['Amount'])
        else:
            tif_transfers_dict[row['TIF']] = int(row['Amount'])

    tif_transfers = pd.DataFrame([(k,v) for k,v in tif_transfers_dict.items()], columns=['TIF', 'Amount'])
    tif_bounds.rename(columns={"name":"TIF"}, inplace=True)
    merged = tif_bounds.merge(tif_transfers, on='TIF', how='left')
    # Create a base layer for the interactive heatmap for census tracts
    fig = px.choropleth_mapbox(
        merged,
        geojson=merged.geometry,
        locations=merged.index,
        color='Amount',  # Color based on population
        color_continuous_scale=["blue", "red"],  # Choose a color scale
        # range_color=[min(tif_transfers['Amount']), max(tif_transfers['Amount'])],  # Set color range explicitly
        range_color=[-7000000, 14000000],  # Set color range explicitly
        mapbox_style="carto-positron",
        center={"lat": merged.centroid.y.mean(), "lon": merged.centroid.x.mean()},
        zoom=10,
        opacity=0.5,
        labels={'TIF'},
        hover_name='TIF',  # Display community area name on hover
        title=year
    )
    fig.show()
```
